```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# https://www.kaggle.com/datasets/lakshmi25npathi/imdb-dataset-of-50k-movie-reviews

# uploading and reading csv file
df = pd.read_csv('/content/IMDB Dataset.csv')
df.sample(10)
```

```python
# checking the dimensions of the dataset
df.shape
```

```python
# checking the null values in dataset
df.isnull().sum()
```

```python
# checking the duplicate values in dataset
print('Duplicate rows:',df.duplicated().sum())
```

```python
# Droping duplicate rows
df = df.drop_duplicates()
```

### Converting Text to Numbers

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['review'])
y = df['sentiment']
```

```python
# It shows total reviews with the list of unique words in all reviews
X.shape
```

```python
# corosponsing output of each review
y.shape
```

### Spliting dataset in for training and testing:

```python
# importng libraries for train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,)
```

### Training the model

```python
# For text classification
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

### Model Evaluation

```python
# For Accuracy Score Evaluation:
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print('Accuracy:',accuracy_score(y_test,y_pred))
```

### Testing by random reviews:

```python
review1 = ['I hated the movie']
review2 = ['I was amazing']
review3 = ['It was so so']


review_vector = vectorizer.transform(review2)
print('Review 01 - Prediction:',model.predict(review_vector))
```
